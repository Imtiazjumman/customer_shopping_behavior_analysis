```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.describe(include='all')  # Describe the dataset to get an overview of the data, including count, mean, std, min, max, and unique values for each column.
```

```python
df.isnull().sum()  # Check for missing values in the dataset. This will return the count of missing values for each column.
```

```python
df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')  # Standardize column names by converting them to lowercase and replacing spaces with underscores.
```

```python
df= df.rename(columns={'purchase_amount_(usd)': 'Purchase amount'})
```

```python
df.columns
```

```python
#create a column by age group
df['Age Group'] = pd.cut(df['age'], bins=[0, 25, 35, 50, 100], labels=['Young', 'Adult', 'Middle-aged', 'Senior'])
```

```python
df[['age', 'Age Group']].head()  # Display the first few rows of the 'age' and 'Age Group' columns to verify the new column has been created correctly.
```

```python
# Create a column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Annually': 365,
    'Bi-Weekly': 14,
    'Quarterly': 90,
    'Every 3 Months': 90
}
df['Purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['Purchase_frequency_days','frequency_of_purchases']].head(10)  # Display the first 10 rows of the 'Purchase_frequency_days' and 'frequency_of_purchases' columns to verify the new column has been created correctly.
```

```python
cols = [c for c in ['discount_applied', 'promo_code_used'] if c in df.columns]
if cols:
    df[cols].head(10)
else:
    print("Neither 'discount_applied' nor 'promo_code_used' exist in df.columns")
```

```python
# Safely check equality only if 'promo_code_used' exists
same_promo_discount = False
if 'promo_code_used' in df.columns:
    same_promo_discount = (df['discount_applied'] == df['promo_code_used']).all()
same_promo_discount
```

```python
df = df.drop(columns=['promo_code_used'], errors='ignore')  # Safe drop if the column exists
```
